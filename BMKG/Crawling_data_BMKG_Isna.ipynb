```python
with urlopen(link) as f:
    tree = ET.parse(f)
root = tree.getroot()
area=[]
```

```python
for movie in root.findall("./forecast/area"):
    area.append(movie.attrib)
    print(movie.attrib)
```

```python
df_area = pd.DataFrame(area)
```

```python
for movie in root.findall("./forecast/area/[@id='"+df_area['id'][0]+"']/parameter"):
    print(movie.attrib)
    parameter.append(movie.attrib)
```

```python
for movie in root.findall("./forecast/area/[@id='"+df_area['id'][0]+"']/parameter"):
    movie.attrib['id_utama'] = df_area['id'][0]
    print(movie.attrib)
```

```python
parameter.append(movie.attrib)
```

```python
df_parameter = pd.DataFrame(parameter)
```

```python
time_range = []
```

```python
for x in range(len(df_parameter)):
    for movie in root.findall("./forecast/area/[@id='"+df_parameter['id_utama'][x]+"']/parameter[@id='"+df_parameter['id'][x]+"']/"):
        movie.attrib['id_utama']=df_parameter['id_utama'][x]
        movie.attrib['parameter_id']=df_parameter['id'][x]
        time_range.append(movie.attrib)
        print(print(movie.attrib))
df_timerange = pd.DataFrame(time_range)
```

```python
final = []
for x in range(len(df_timerange)):
    for movie in root.findall("./forecast/area/[@id='"+df_timerange['id_utama'][x]+"']/parameter[@id='"+df_timerange['parameter_id'][x]+"']/timerange[@datetime='"+df_timerange['datetime'][x]+"'][@type='"+df_timerange['type'][x]+"'][@h='"+str(df_timerange['h'][x])+"']/"):
        movie.attrib['id']=df_timerange['id_utama'][x]
        movie.attrib['parameter_id']=df_timerange['parameter_id'][x]
        movie.attrib['datetime']=df_timerange['datetime'][x]
        movie.attrib['type']=df_timerange['type'][x]
        movie.attrib['h']=df_timerange['h'][x]
        movie.attrib['day']=df_timerange['day'][x]
        movie.attrib['value']=movie.text +' '+movie.attrib['unit']
        print(movie.attrib)
        final.append(movie.attrib)
```

```python
akhir=[]
area=[]
for movie in root.findall("./forecast/area"):
    area.append(movie.attrib)
    print(movie.attrib)
df_area = pd.DataFrame(area)
for xx in range(len(df_area)):

        parameter = []
        for movie in root.findall("./forecast/area/[@id='"+df_area['id'][xx]+"']/parameter"):
            movie.attrib['id_utama'] = df_area['id'][xx]
            print(movie.attrib)

            parameter.append(movie.attrib)
        #     x=x+1
        # print(x)

        df_parameter = pd.DataFrame(parameter)


        time_range = []
        for x in range(len(df_parameter)):
            for movie in root.findall("./forecast/area/[@id='"+df_parameter['id_utama'][x]+"']/parameter[@id='"+df_parameter['id'][x]+"']/"):
                movie.attrib['id_utama']=df_parameter['id_utama'][x]
                movie.attrib['parameter_id']=df_parameter['id'][x]
                time_range.append(movie.attrib)
                print(print(movie.attrib))
        df_timerange = pd.DataFrame(time_range)

        final = []
        for x in range(len(df_timerange)):
            for movie in root.findall("./forecast/area/[@id='"+df_timerange['id_utama'][x]+"']/parameter[@id='"+df_timerange['parameter_id'][x]+"']/timerange[@datetime='"+df_timerange['datetime'][x]+"'][@type='"+df_timerange['type'][x]+"'][@h='"+str(df_timerange['h'][x])+"']/"):
                movie.attrib['id']=df_timerange['id_utama'][x]
                movie.attrib['parameter_id']=df_timerange['parameter_id'][x]
                movie.attrib['datetime']=df_timerange['datetime'][x]
                movie.attrib['type']=df_timerange['type'][x]
                movie.attrib['h']=df_timerange['h'][x]
                movie.attrib['day']=df_timerange['day'][x]
                movie.attrib['value']=movie.text +' '+movie.attrib['unit']

                print(movie.attrib)
                final.append(movie.attrib)

        df_final = pd.DataFrame(final)
        akhir.append(df_final)
```

```python
final_akhir = pd.concat(akhir)
```

```python
final_akhir
```

```python
final_akhir.merge(df_area)
```

```python
pd.merge(final_akhir, df_area, on="id", how="left")
```

```python
df = pd.merge(final_akhir, df_area, on="id", how="left")
```

```python
df['ket_weather'] = np.where(df['parameter_id']=='weather', df['value'].apply(lambda x: x.split(" ", 1)[0]).map(dict_cuaca), np.NaN)
df['ket_wd'] = np.where(df['parameter_id']=='wd', np.where(df['value'].apply(lambda x: x.split(" ", 1)[1]) == 'CARD', df['value'].apply(lambda x: x.split(" ", 1)[0]).map(dict_arah_angin),np.NaN),np.NaN)
df['updated_date'] = datetime.now()
```
