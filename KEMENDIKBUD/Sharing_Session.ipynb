```python
#melakukan crawling untuk mendpatkan value dropdown yang nantinya akan digunakan sebagai input ajax
wiki_link = "https://bansm.kemdikbud.go.id/akreditasi"
html = urlopen(wiki_link).read()
soup = BeautifulSoup(html, 'html.parser')
```

###  Parameter Provinsi

```python
#mendapatkan nama provinsi
prov = soup.find("select", {"id": "id_provinsi"}).findAll('option')
values = [o.get("value") for o in prov]
prov = values[1:]
prov
prov_name = soup.find("select", {"id": "id_provinsi"})
nama_prov = []
for value in prov_name.stripped_strings:
    nama_prov.append(value)
#     print(value)
nama_prov = nama_prov[1:]
```

### Mendapatkan Semua id kabupaten dari tiap Provinsi

```python

# from json import loads
# import certifi
# from collections import deque
all_kabid_array =[]
all_kabname_array =[]
for xx in range(len(prov)):
    print(prov[xx])

    # urlopen(request, context=ssl.create_default_context(cafile=certifi.where()))
    session = Session()

    # HEAD requests ask for *just* the headers, which is all you need to grab the
    # session cookie
    session.head('https://bansm.kemdikbud.go.id/',verify=False)
#     session.cert=certifi.where()
    response = session.post(
        url='https://bansm.kemdikbud.go.id/akreditasi/getKabKot',
        data={
            'idprov':prov[xx]
        },
        headers={
            'Referer': 'https://bansm.kemdikbud.go.id/akreditasi'
        }
    )
    # dic = loads(response.text)
    # print(response.json)
    # pd.DataFrame.from_dict(response.json)
    output = ""+response.text+""
    output = "<select>"+output+"</select>" 
    web_soup = soup(output)
    reviews = web_soup.find('select')
    delimiter = str(web_soup.contents[0]).split('"')
    id_kab = []
    for x in range(len(str(web_soup.contents[0]).split('"'))):
        if x%2 == 1:
            print(prov[xx]+'==')
            print(delimiter[x])
            id_kab.append(delimiter[x])
        
    all_kabid_array.append(id_kab[1:])
    nama_kabkot = []
    for value in reviews.stripped_strings:
        nama_kabkot.append(value)
        print(value)
    all_kabname_array.append(nama_kabkot[1:])
#         print(delimiter[x])
```
