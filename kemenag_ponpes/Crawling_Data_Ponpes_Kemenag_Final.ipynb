```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://ditpdpontren.kemenag.go.id/pdpp//loadpp?loadpp=&page=1"
r = requests.get(url)
request = r.content
soup = BeautifulSoup(request, "html.parser")
```

```python
all_prov = soup.find("select", class_="provinsi form-control").findAll("option")

list_prov = []
for i in range(len(all_prov)):
    id_provinsi = all_prov[i]['value']
    nama_provinsi = all_prov[i].text
#     print(nama_provinsi)
    prov_dict = {
          'id_provinsi': id_provinsi,
          'nama_provinsi': nama_provinsi
        }
    print(prov_dict)
    list_prov.append(prov_dict)
Lprovinsi = pd.DataFrame(list_prov)
```

```python
getkab = "https://ditpdpontren.kemenag.go.id/pdpp/getkabupaten/12"
r = requests.get(getkab)
request = r.content
soup = BeautifulSoup(request, "html.parser")

soup = [soup]
# print(soup)
Lkab = soup[0].text[1:-1].strip().replace('"','').split(',')
Lkab
```

```python
list_kab = []

for ii in range(len(Lkab)):
    list_1= Lkab[ii].split(":")
    l_dict = {
          'id_kabupaten': list_1[0],
          'nama_kabupaten': list_1[1]
        }
    print(l_dict)
    list_kab.append(l_dict)
list_kabupaten = pd.DataFrame(list_kab)
```

```python
list_kabupaten
```

```python
prov_kab=[]

for z in range(len(Lprovinsi)):
    if(len(str(Lprovinsi['id_provinsi'][z]))<1):
        continue
    else:
        u_link = "https://ditpdpontren.kemenag.go.id/pdpp/getkabupaten/"+str(Lprovinsi['id_provinsi'][z])
        print(u_link)
        r = requests.get(u_link)
        request = r.content
        soup = BeautifulSoup(request, "html.parser")
        soup = [soup]
        isi_kab = []
        Lkab=soup[0].text[1:-1].strip().replace('"','').split(',')
#         print(Lkab)
        for zz in range(len(Lkab)):
            list_1= Lkab[zz].split(":")
            dict_provkab = {
                   'id_provinsi': Lprovinsi['id_provinsi'][z],
                    'nama_provinsi': Lprovinsi['nama_provinsi'][z],
                    'id_kabupaten': list_1[0],
                    'nama_kabupaten': list_1[1]
                }
            print(dict_provkab)
            isi_kab.append(dict_provkab)
    isi_kab = pd.DataFrame(isi_kab)
    prov_kab.append(isi_kab)
```

```python
prov_kab = pd.concat(prov_kab)
prov_kab = prov_kab.reset_index()
prov_kab
```

```python
prov_kab = prov_kab[prov_kab['index']!=0]
prov_kab = prov_kab.reset_index()
prov_kab
```
